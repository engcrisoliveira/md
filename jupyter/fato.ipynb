#DimAluno

```python
#Carregando a tabela dimAluno
dimAluno = pd.read_csv('df_aluno.csv', usecols=['ID_ALUNO','RACA','PAIS_ORIGEM','GENERO','ANO_NASCIMENTO'])
dimAluno.head()

#Inserindo a coluna idade
dimAluno['IDADE']= 2019 - dimAluno['ANO_NASCIMENTO']
dimAluno.head()

#Selecionando as colunas do dataframe dimAluno
dimAluno = dimAluno[['ID_ALUNO','RACA','PAIS_ORIGEM','GENERO','IDADE']]
dimAluno.head()
```

```python
#Join com arquivo csv com key e id_aluno
dimAluno = pd.merge(dimAluno, key_id_dimaluno, how = 'inner', on = 'id_aluno')
dimAluno.head()
```

```python
#Renomeando as colunas
dimAluno = dimAluno.rename(columns={
    'key':'DimAluno_key',
    })
dimAluno.head()
```

```python
dimAluno = dimAluno[['DimAluno_key','id_aluno','raca','origem','genero','idade']]
dimAluno.head()
```
