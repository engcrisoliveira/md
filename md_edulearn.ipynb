#Entidade Aluno

```python
df_aluno = pd.read_csv('alunodeficiente_2019.csv', sep=',', usecols=['CO_IES','TP_COR_RACA','NU_ANO_NASCIMENTO','ID_ALUNO','CO_PAIS_ORIGEM','TP_SEXO'])
df_aluno.head()
```

```python
#Renomeando as colunas
df_aluno = df_aluno.rename(columns={
    'CO_IES':'CODIGO_IES',
    'ID_ALUNO':'CODIGO_INEP',
    'TP_COR_RACA':'RACA',
    'TP_SEXO':'GENERO',
    'CO_PAIS_ORIGEM':'PAIS_DE_ORIGEM',
    })
df_aluno.head()
```

```python
#Substituindo o código pela raça
raca = {
    '0':'Aluno não quis declarar cor/raça',   
    '1':'Branca',
    '2':'Preta',
    '3':'Parda',
    '4':'Amarela',
    '5':'Indígena',
    '9':'Não dispõe da informação',
}
df_aluno = df_aluno.astype({'RACA':str})
df_aluno.RACA = df_aluno.RACA.map(raca)
df_aluno.head()
```

```python
#Substituindo código por gênero
genero = {
    '1':'Feminino',
    '2':'Masculino',
}
df_aluno = df_aluno.astype({'GENERO':str})
df_aluno.GENERO = df_aluno.GENERO.map(genero)
df_aluno.head()
```

```python
data = datetime.date.today()
ano_atual = date.strftime("%Y")
df_aluno['ANO_ATUAL'] = ano_atual
df_aluno.head()
```

```python
#Inserindo a coluna idade do aluno
df_aluno = df_aluno.astype({'ANO_ATUAL':int})
df_aluno = df_aluno.astype({'NU_ANO_NASCIMENTO':int})
df_aluno['IDADE'] = df_aluno['ANO_ATUAL'] - df_aluno['NU_ANO_NASCIMENTO']
df_aluno.head()
```

```python
#Verificando se há CODIGO_INEP duplicaado
df_aluno['CODIGO_INEP'].duplicated()
```

Observação: Como há códigos do inep duplicados para alunos com dados diferentes, resolvemos não considerar essa coluna como identificadora e colocar uma id AI no banco de dados como chave primaria do aluno.

```python
#Excluindo a coluna código inep
df_aluno = df_aluno.drop(columns=['CODIGO_INEP'])
```

```python
#Para referenciar o código da IES com o id da IES do banco de dados
id_ies = pd.read_csv('tabela_com_id_ies.csv')
id_ies.head()
```

```python
#Dando um join com o código da IES
df_aluno = df_aluno.join(id_ies[['id_ies', 'cod']].set_index('cod'), on='CODIGO_IES')
df_aluno.head()
```

```python
paises = df_aluno['PAIS_DE_ORIGEM'].unique()
paises = sorted(paises)
paises
```

```python
df_aluno.dtypes
```

```python
#Substituindo o código pelo país de origem
pais = {
    24:'Angola',
    32:'Argentina',
    64:'Butão',
    68:'Bolívia',
    76:'Brasil',
    108:'Burundi',
    120:'Camarões',
    132:'Cabo Verde',
    152:'Chile',
    156:'China',
    170:'Colômbia',
    178:'República do Congo',
    180:'República Democrática do Congo',
    192:'Cuba',
    204:'Benim',
    218:'Equador',
    250:'França',
    254:'Guiana Francesa',
    276:'Alemanha',
    332:'Haiti',
    380:'Itália',
    392:'Japão',
    410:'Coreia do Sul',
    508:'Moçambique',
    528:'Holanda',
    558:'Nicarágua',
    600:'Paraguai',
    604:'Peru',
    608:'Filipinas',
    620:'Portugal',
    624:'Guiné-Bissau',
    626:'Timor-Leste',
    678:'São Tomé e Príncipe',
    686:'Senegal',
    702:'Cingapura',
    724:'Espanha',
    752:'Suécia',
    756:'Suíça',
    760:'Síria',
    840:'Estados Unidos',
    858:'Uruguai',
    862:'Venezuela',
    999:'Outra Nacionalidade', 
}
df_aluno = df_aluno.astype({'PAIS_DE_ORIGEM':int})
df_aluno.PAIS_DE_ORIGEM = df_aluno.RACA.map(pais)
df_aluno.head()
```

```python
#Organizando as colunas do dataframe
df_aluno = df_aluno.rename(columns={
    'id_ies':'ID_IES'})
#df_aluno['ID_ALUNO'] = df_aluno.index
df_aluno = df_aluno[['ID_ALUNO','RACA','IDADE','PAIS_DE_ORIGEM','GENERO','ID_IES']]
df_aluno.head()
```

# Tabela de recursos

```python
import pandas as pd
df_recursos = pd.read_csv('curso_e_ies_2019.csv', usecols=['CODIGO_CURSO','RECURSOS_COMUNICACAO','RECURSOS_INFORMATICA'])
df_recursos.head()
```

```python
df_recursos = df_recursos.join(df_cursos[['ID','CODIGO_CURSO']].set_index('CODIGO_CURSO'), on='CODIGO_CURSO')
df_recursos.head()
```

```python
#Renomeando as colunas
df_recursos = df_recursos.rename(columns={
    'ID':'ID_CURSO'})
df_recursos.head()
```

```python
df_comunicacao = df_recursos.loc[df_recursos['RECURSOS_COMUNICACAO'] == 1]
df_comunicacao['ID_RECURSO'] = 1
df_comunicacao = df_comunicacao[['ID_CURSO','ID_RECURSO']]
df_comunicacao.head()
```

```python
df_informatica = df_recursos.loc[df_recursos['RECURSOS_INFORMATICA'] == 1]
df_informatica['ID_RECURSO'] = 2
df_informatica = df_informatica[['ID_CURSO','ID_RECURSO']]
df_informatica.head()
```

```python
df_recurso = pd.concat([df_comunicacao, df_informatica])
df_recurso.head()
```

```python
df_recurso.tail()
```

```python
df_recurso.to_csv('tabela_curso_recurso.csv', sep=';')
```
