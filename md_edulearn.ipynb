##Aluno

```python
df_aluno = pd.read_csv('alunodeficiente_2019.csv', sep=',', usecols=['CO_IES','TP_COR_RACA','NU_ANO_NASCIMENTO','NU_MES_NASCIMENTO','NU_DIA_NASCIMENTO','ID_ALUNO','CO_PAIS_ORIGEM','TP_SEXO'])
df_aluno.head()
```

```python
#Inserindo a coluna data de nascimento
df_aluno['IDADE']= 2021 - df_aluno['NU_ANO_NASCIMENTO']
df_aluno.head()
```

```python
#Renomeando as colunas
df_aluno = df_aluno.rename(columns={
    'CO_IES':'CODIGO_IES',
    'ID_ALUNO':'CODIGO_INEP',
    'TP_COR_RACA':'RACA',
    'TP_SEXO':'GENERO',
    'CO_PAIS_ORIGEM':'PAIS_DE_ORIGEM',
    })
df_aluno.head()
```

```python
#Substituindo código por país de origem
origem = {
    '76':'Brasil',
}
df_aluno = df_aluno.astype({'PAIS_DE_ORIGEM':str})
df_aluno.PAIS_DE_ORIGEM = df_aluno.PAIS_DE_ORIGEM.map(origem)
df_aluno.head()
```

```python
#Substituindo o código pela raça
raca = {
    '0':'Aluno não quis declarar cor/raça',   
    '1':'Branca',
    '2':'Preta',
    '3':'Parda',
    '4':'Amarela',
    '5':'Indígena',
    '9':'Não dispõe da informação',
}
df_aluno = df_aluno.astype({'RACA':str})
df_aluno.RACA = df_aluno.RACA.map(raca)
df_aluno.head()
```

```python
#Substituindo código por gênero
genero = {
    '1':'Feminino',
    '2':'Masculino',
}
df_aluno = df_aluno.astype({'GENERO':str})
df_aluno.GENERO = df_aluno.GENERO.map(genero)
df_aluno.head()
```

```python
id_ies = pd.read_csv('tabela_com_id_ies.csv')
id_ies.head()
```

```python
id_ies = id_ies.rename(columns={
    'cod':'CODIGO_IES',
    'id_ies':'ID_IES',
})
id_ies.head()
```

```python
#Dando um merge com o código da IES
df_aluno = pd.merge(id_ies, df_aluno, how = 'inner', on = 'CODIGO_IES')
df_aluno.head()
```

```python
#Preparando a tabela aluno
#df_aluno = df_aluno.drop_duplicates()
df_aluno = df_aluno[['RACA','IDADE','CODIGO_INEP','PAIS_DE_ORIGEM','GENERO','ID_IES']]
df_aluno
```

###Número de recursos, alunos matriculados e concluintes

```python
#Importando os dados de cursos
cursos_recursos = pd.read_csv('df_curso.csv', usecols=['CODIGO_CURSO','NUM_ALUNOS_MATRICULADOS','NUM_ALUNOS_CONCLUINTES','MATERIAL_DIGITAL','MATERIAL_AMPLIADO','MATERIAL_TATIL','MATERIAL_IMPRESSO','MATERIAL_AUDIO','MATERIAL_BRAILLE','MATERIAL_LIBRAS','DISCIPLINA_LIBRAS','TRADUTOR_LIBRAS','GUIA_INTERPRETE','RECURSOS_COMUNICACAO','RECURSOS_INFORMATICA'])
cursos_recursos.head(3)
```

```python
#Calcular a quantidade de recursos de acessibilidade ofertados por cada curso
def contar_recursos(row):
  return row['MATERIAL_DIGITAL'] + row['MATERIAL_AMPLIADO'] + row['MATERIAL_TATIL'] + row['MATERIAL_IMPRESSO'] + row['MATERIAL_AUDIO'] + row['MATERIAL_BRAILLE'] + row['MATERIAL_LIBRAS'] + row['DISCIPLINA_LIBRAS'] + row['TRADUTOR_LIBRAS'] + row['GUIA_INTERPRETE'] + row['RECURSOS_COMUNICACAO'] + row['RECURSOS_INFORMATICA']

cursos_recursos['RECURSOS'] = cursos_recursos.apply(contar_recursos, axis=1)
cursos_recursos.head()
```

```python
#Substituindo NaN por zero
cursos_recursos['RECURSOS'].fillna(0, inplace=True)
cursos_recursos.head(3)
```

```python
cursos_recursos = cursos_recursos[['CODIGO_CURSO','NUM_ALUNOS_MATRICULADOS','NUM_ALUNOS_CONCLUINTES','RECURSOS']]
cursos_recursos.head()
```

```python
#Exportando o dataframe para um arquivo csv
cursos_recursos.to_csv('fatocursos_dw.csv', sep=',')
```
